```python
import tushare as ts
df=ts.shibor_data('2017').append(ts.shibor_data('2018'))
df.sort_values('date',ascending=True,inplace=True)
df.plot(x='date',figsize=[10,6])
```

```python
import tushare as ts
df=ts.realtime_boxoffice()
df['sumBoxOffice']=df['sumBoxOffice'].astype('float')
df.movieDay=df.movieDay.astype('int')
df['boxPerDay']=df.sumBoxOffice/df.movieDay
df.sort_values('boxPerDay',ascending=False)
```

```python
import tushare as ts

df=ts.get_money_supply_bal()
df['m2']=df['m2'].astype('float')
df.year=df.year.astype('int')
df['m2_rate']=(df.m2-df['m2'].shift(-1))/df['m2'].shift(-1)
df
```

```python
df.plot(x='year',y='m2',figsize=[20,10])
```

```python
import pandas as pd
df1=ts.get_gdp_year()
df2=pd.merge(df,df1,on='year')
df2.gdp=df2.gdp.astype('float')
df2['m2_gdp']=df2.m2/df2.gdp
df2['m2_rate']=df2['m2_rate'].map(lambda x:format(x,'.2%'))
df2.iloc[:,[0,1,9,10,-1]]
```
